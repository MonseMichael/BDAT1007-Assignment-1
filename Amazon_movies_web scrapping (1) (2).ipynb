```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://editorial.rottentomatoes.com/guide/best-movies-on-amazon-prime-right-now/"
page=requests.get(url)

content = page.content
soup = BeautifulSoup(content, "lxml")
print(soup)
```

```python
title=[]

title_ele=soup.find_all(class_="article_movie_title")
for x in tit_ele:
    title.append(x.text.partition('(')[0][1:])
print(title)
```

```python
year=[]

year_ele=soup.find_all(class_="subtle start-year")
for x in year_ele:
    year.append(x.text[1:5])
print(year)
```

```python
rating=[]

rating_ele=soup.find_all(class_="tMeterScore")
for x in rating_ele:
    rating.append(x.text)
print(rating)
```

```python
rank=[]

rank_ele=soup.find_all(class_="countdown-index")
for x in rank_ele:
    rank.append(x.text[1:4])
print(rank)
```

```python
director=[]

director_ele=soup.find_all(class_="info director")
for x in director_ele:
    director.append(x.text[14:])
print(director)
```

```python
final_array=[]

for x,y,z,a,b in zip(tit,year,rating,rank,director):
    final_array.append({"Title":x,"Year":y,"Rating":z,"Rank":a,"Director":b})
data=pd.DataFrame(final_array)

data
```
